```python
import geopandas as gpd

import matplotlib.pyplot as plt

#import shapefile and assign to a variable
places = gpd.read_file("GH4/ghana_location.shp")
```

```python
#find the datatyple
type(places)

#view first 5 rows of the data
#places.head()

#view the column/heading names
#places.columns
```

```python
#plot map with color jet, specify column to categorise color, and set the map size
places.plot( cmap = "jet", column ="NAME", figsize = (10,10))
#len(dev_sites)
```

```python
AOI = gpd.read_file("GH4/10_regions.shp")
#AOI.head()
AOI.plot(cmap = "jet", column = "NAME_1", figsize = (10,10))
```

```python
#Take out/Subset Eastern Region from the entire regional data
eastern = AOI[AOI["NAME_1"] == "Eastern"]
```

```python
eastern.plot()
```

```python
#Use eastern regions to clip the data
places_clipped = gpd.clip(places, eastern)
places_clipped.plot()
```

```python
#plot both shapefiles together
fig, ax = plt.subplots(1)

eastern.plot(ax=ax, color = "brown")
places_clipped.plot(ax=ax, color = "green")
```

```python
#intersection
places_in_eastern = gpd.overlay(places, eastern, how = "intersection")
places_in_eastern.plot(figsize = (10,10), color = "black")
```
